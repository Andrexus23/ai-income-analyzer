1. Насколько выше доход у фрилансеров, принимающих оплату в криптовалюте, по сравнению с другими способами оплаты?

```python
income_by_payment = df.groupby('Payment_Method')['Earnings_USD'].agg([
            'mean', 'median', 'count', 'std'
]).round(2)
income_by_payment
```

```python
income_by_payment.to_dict()
```

2. Как распределяется доход фрилансеров в зависимости от региона проживания?

```python
# 2. Распределение дохода по регионам
df.groupby('Client_Region')['Earnings_USD'].agg([
    'mean', 'median', 'count', 'std'
]).round(2)
```

```python
df.groupby('Client_Region')['Earnings_USD'].agg([
    'mean', 'median', 'count', 'std'
]).round(2)
```

```python
result_dict = (df.groupby('Client_Region')['Earnings_USD']
                .agg(['mean', 'count', 'std'])
                .round(2)
                .reset_index()
                .to_dict('list'))

final_dict = {
    'Client_Region': result_dict['Client_Region'],
    'Earnings_USD_mean': result_dict['mean'],
    'Earnings_USD_count': result_dict['count'],
    'Earnings_USD_std': result_dict['std']
}
final_dict
```

3. Какой процент фрилансеров, считающих себя экспертами, выполнил менее 100 проектов?

```python
experts = df[(df['Experience_Level'] == 'Expert')]
# experts = experts[['Experience_Level', 'Job_Completed']]
len(experts)
```

```python
from typing import Dict, List
def get_freelancers_where(condition: str, page_content: str, fields: List = []):
    relevant_part: Dict = df.query(condition)
    if fields:
        relevant_part = relevant_part[fields]
    return {
        'page_content': f'{page_content}: {str(relevant_part.to_dict())}',
    }

get_freelancers_where(
    'Experience_Level == "Expert"', 
    page_content='Фрилансеры, считающие себя экспертами и их количество проектов (id, Job_Completed)',
    fields=['Job_Completed'],
)
```

```python
# optional
round((len(experts[experts['Job_Completed'] < 100]) / len(experts)) * 100, 2)
```
